```python
import pandas as pd
import seaborn as sns
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
df = pd.read_csv('Selim_Hossain_Data.csv')
df.head(5)
```

```python
df.dtypes
```

```python
df['EST'] = pd.to_datetime(df['EST'])
df['EST']
```

```python
df.dtypes
```

```python
df.set_index('EST')
```

#### line plot with one variable

```python
fig = plt.figure(figsize= (12, 3))
sns.set(style='darkgrid')
ax = sns.lineplot(x= df['EST'], y= df['Temperature(C)'], color= 'red')
ax.set_title('Time vs Temperature')
ax.set_xlabel('Time')
```

####  Line plot with moving average

```python

#moving average

mov_avg = df['Temperature(C)'].rolling(10).mean()
plt.figure(figsize= (12, 4))

ax = sns.lineplot(x= df['EST'], y= df['Temperature(C)'], color= 'blue', label= 'Without Average')
ax = sns.lineplot(x = df['EST'], y = mov_avg, color= 'red', label= 'Moving Average')
ax.set_title('Moving Average with Time', fontsize= 20)
ax.legend()
ax.set_xlabel('Time')
```

#### Line plot of two variables with markers

```python
# take the figure size
fig = plt.figure(figsize=(10, 5))

x= df['EST']
y1= df['Temperature(C)']
y2=df['DewPoint(F)']
y3= df['Humidity(Kg/m3)']


#As we have defined only one axis (ax) everything is plotted on this axis. 
ax = sns.lineplot(x, y1, marker='D', linestyle='-',markersize=8, linewidth=2, label='Temperature')

ax = sns.lineplot(x, y2, marker='o', markersize=8, linestyle='-',linewidth=2, label='Dewpoint')

ax = sns.lineplot(x, y3, marker='^', linestyle='-',markersize=8, linewidth=2, label='Humidity')


#We insert a legend 
ax.legend()

#Labels and title
plt.xlabel('Time')
plt.ylabel('Count')
plt.title('Temperature, Dewpoint and Humidity', color= 'red',fontsize=15)
plt.show()
```

####  Plot with two variables and running mean

```python
moving_average= df['Temperature(C)'].rolling(10).mean()
moving_average2= df['DewPoint(F)'].rolling(10).mean()

#Definition of the subplots
plt.figure(figsize= (10,6))

ax = sns.lineplot(x = df['EST'], y= df['Temperature(C)'], marker='o', markersize=6, color='blue',linestyle='None', label='Temperature')
ax = sns.lineplot(x= df['EST'], y= moving_average, marker='o', markersize=6,linewidth=1.5, color='m', label='Moving average Temp')


ax = sns.lineplot(df['EST'], y= df['DewPoint(F)'], marker='D', markersize=6, color='green', linestyle='None', label='Dewpoint')
ax = sns.lineplot(df['EST'], y= moving_average2, marker='D', markersize=6,linewidth=1.5, color='cyan', label='Moving average Dewpoint')

#Set legend, labels and title
ax.legend()
ax.set_xlabel('Time')
ax.set_ylabel('Accumulation')
ax.set_title('Accumulation Vs Time', color= 'darkorange' ,fontsize=(15))
plt.show()
```

####  Plot with 3 variables subplotted

```python
#Assigning the variables
x = df['EST']
y1 = df['Temperature(C)']
y2 = df['DewPoint(F)']
y3 = df['Humidity(Kg/m3)']

#Now we use 3 axes (ax, ax1, ax2) and we define the settings for the axes block by block
fig, (ax, ax1, ax2) = plt.subplots(3, 1, sharex=True, figsize=(16,8))

#Change color, linestyle, linewidth, markerstyle and transparency (alpha)
ax.plot(x, y1, color='mediumaquamarine', marker='o', alpha=1, linestyle='-')
#Change title and fontsize
ax.set_title('Temperature', fontsize=(12))
ax.set_ylabel('Accumulation')

#Second plot
ax1.plot(x, y2, color='steelblue', marker='d', alpha=1, linestyle='-')
ax1.set_title('DewPoint', fontsize=(12))
ax1.set_ylabel('Accumulation')

#Third plot
ax2.plot(x, y3, color='palevioletred', marker='D', alpha=1, linestyle='-')
ax2.set_title('Humidity', fontsize=(12))
ax2.set_ylabel('Accumulation')
plt.show()
```

####  Plot with filled areas to reveal information on positive and negative changings

```python
#Assigning the variables
x = df['EST']
y1 = df['Temperature(C)']
y2 = df['DewPoint(F)']
y3 = df['Humidity(Kg/m3)']

fig, (ax, ax1) = plt.subplots(2, 1, sharex=True, figsize=(16,8))


#Settings for upper plot
ax.plot(x, y1, x, y2, color='grey')
ax.fill_between(x, y1, y2, where=y2 >= y1, facecolor='lightblue', interpolate=True)
ax.fill_between(x, y1, y2, where=y2 <= y1, facecolor='lightcoral', interpolate=True)
ax.set_title('Differences in accumulation,', fontsize=(16))
ax.set_ylabel('Accumulation')

#Settings for lower plot
ax1.plot(x, y2, x, y3, color='grey')
ax1.fill_between(x, y2, y3, where=y3 >= y2, facecolor='lightblue', interpolate=True)
ax1.fill_between(x, y2, y3, where=y3 <= y2, facecolor='lightcoral', interpolate=True)
ax1.set_title(' revealing positive (blue) and negative (red) accumulation in comparison.', fontsize=(16))
ax1.set_ylabel('Accumulation')
plt.show()
```
