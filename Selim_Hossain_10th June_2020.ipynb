```python
import pandas as pd
import numpy as np
import seaborn as sns
import scipy as sp
import matplotlib.pyplot as plt 
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
df = pd.read_csv('Selim_Hossain_data.csv')
df
```

```python
#Histogram or distribution plot

plt.hist(df['SeaDistance(Km)'],
              facecolor = "blue",
              edgecolor = "lightblue",
              bins = 10,cumulative = True)

plt.hist(df['Temperature(C)'],facecolor = "red",
              edgecolor = "red",
              bins = 10,alpha = 0.7, range= (5, 50))

plt.xlabel('Number of bins')
plt.ylabel('Frequency')
plt.title('Two histogram Graph')
plt.show()
```

```python
#Pie plot.Normally use for categorical data

labels = ['Temp', 'Dewpoint', 'Humidity', 'Sea Level', 'Windspeed']
sizes = [df['Temperature(C)'], df['DewPoint(F)'], df['Humidity(Kg/m3)'], df['Sea Level Pressure(Pa)'], df['WindSpeed(MPH)']]

freq_density = sp.stats.itemfreq(df["Events"])
df1 = pd.DataFrame(freq_density, columns=['Density_SI', 'Frequency'])
labels= df1['Density_SI']
sizes = df1['Frequency']
#print(df1)

plt.pie(sizes, labels=labels,autopct='%1.1f%%',shadow=True, startangle=90)
plt.title('Density functiion With label')
plt.show()
```

```python
#Line plot. Line plot normally use for the graph changing with time
fig = plt.figure(figsize= (12, 5))

ax1 = fig.add_axes([0,0, 1,1])
ax2 = fig.add_axes([0.50, 0.7, 0.5, 0.3])

ax1.plot(df['EST'], df['Sea Level Pressure(Pa)'],color= 'b', label= 'Sea Level Pressure')
ax2.scatter(df['EST'], df['DewPoint(F)'], color= 'm',alpha= 0.9)

ax1.legend(loc='upper left')
ax1.set_xlabel('Day')
ax1.set_ylabel('Sel Level Pressure(PA)')
ax1.set_title('Sea label Pressure and Dewpoint')

#ax2 attribute
ax2.set_xlabel('Temperature')
ax2.set_ylabel('Dewpoint')

plt.show()
```

```python
# Define the size of the whole figure (width/height)
fig = plt.figure(figsize =(20,8))

# Add a title
plt.suptitle("Comparison of different types of graph",
            fontsize = 20)

#Create subplot for temperature
ax1 = fig.add_subplot(221) # position of individual subplot (Matrix = 111)
ax1.set_title("Temperature") # Title of subplot
ax1.plot(df['EST'],
         df['Temperature(C)'],
        color = "r") # choose individual colors
plt.axhline(y = 30, linewidth=5, color='m') # Insert horizontal line 

#Create plot Dewpoint
ax2 = fig.add_subplot(222)
ax2.set_title("Dewpoint")
ax2.plot(df['EST'],
         df['DewPoint(F)'],
        color = "b")
plt.axhline(y = 30, linewidth=5, color='m')

#Create plot Humidity(Kg/m3)
ax3 = fig.add_subplot(223)
ax3.set_title("Humidity(Kg/m3)")
ax3.plot(df['EST'],
         df['Humidity(Kg/m3)'],
        color = "g")
plt.axhline(y = 30, linewidth=5, color='m')

#Create plot ax4 Sea Level Pressure(Pa)
ax4 = fig.add_subplot(224)
ax4.set_title("Sea Level Pressure(Pa)")
ax4.plot(df['EST'],
         df['Sea Level Pressure(Pa)'],
        color = "cyan")

plt.axhline(y = 30, linewidth=5, color='m')
plt.show()
```

```python
#Boxplot, Generally,Boxplot use for categorical data

x = df['Temperature(C)']
xx = x[~np.isnan(x)]

y = df['DewPoint(F)']
yy = y[~np.isnan(y)]

z = df['Humidity(Kg/m3)']
zz = z[~np.isnan(z)]

datas = [xx, yy, zz]
colors = ['red', 'green', 'cyan']

ax = sns.boxplot(data= datas, notch=True)
ax.set_title('Boxplot')

#bp = plt.boxplot(datas, patch_artist=True, notch= True)

# for i in range(len(bp['boxes'])):
#     bp['boxes'][i].set(facecolor = colors[i])
#     bp['caps'][2*i + 1].set(color= colors[i])
    
plt.xticks([0, 1, 2], ['Temp', 'Dewpoint', 'Humidity'])
plt.show()
    
```

```python
#Violinplot

vp = plt.violinplot(datas, showmedians= True)

for i in range(len(vp['bodies'])):
    vp['bodies'][i].set(facecolor= colors[i])
plt.xticks([1,2,3], ['Temperature', 'Dewpoint', 'Humidity'])
plt.title('Violinplot')
plt.show()
```

```python
#Stackplot...

#Applying the rolling mean on the data set

rolling_mean1=df['Temperature(C)'].rolling(window=20, min_periods=1).mean()
rolling_mean2=df['DewPoint(F)'].rolling(window=20, min_periods=1).mean()
rolling_mean3=df['Humidity(Kg/m3)'].rolling(window=20, min_periods=1).mean()

x = df['EST'] 
y = np.vstack((rolling_mean1, rolling_mean2, rolling_mean3))

colors = ['red', 'green', 'cyan']
labels = ['Temperature', 'Dewpoint', 'Humidity']

plt.figure(figsize= (12, 6))

plt.stackplot(x,y, labels= labels, colors = colors, edgecolor= 'white')
plt.legend()
plt.tight_layout()
```

```python
#Stem plot normally use for discrete signal

# Create a copy of your data set

stem_df = df.copy()

# Compare two data sets or variables using `diff()`
stem_df[['temp', 'dew']] = df[['Temperature(C)','DewPoint(F)']].diff()

# Define the size of your figure
plt.figure(figsize=(15,11))

# Add a main title and define position(y=) and letter size(fontsize=)
plt.suptitle('Some Temperature and Dewpoint change', y=1.03 , fontsize=20)


# Create first subplot(position matrix)
plt.subplot(311)
plt.stem(df['EST'],
         stem_df['temp'], # y-axis first plot
         markerfmt = 'g_', # Defining the format of the markers at x/y-position -> 'g_' = "green, flat line"
         linefmt = 'b--', # Defining the format of the lines towards x/y-position -> 'b--' = "blue, dashed line"
         basefmt='r-') # Defining the format of the bottom line (by default at position 0) -> 'r-' = "red, solid line"
plt.title('Temperature(C)')


plt.subplot(312)
plt.stem(df['EST'],
         stem_df['dew'], # y-axis second plot
         markerfmt = 'g_',
         linefmt = 'b--',
         basefmt='r-')
plt.title('DewPoint(F)')

# Add a little text box inside the plot, by adding this function to the addressed plot
#plt.text(65, -110, 'Differences in data size between the years, due to the loss of measuring stakes', style='italic',
       # bbox={'facecolor': 'azure', 'alpha': 0.5, 'pad': 10}) # Define th style of the box

# Improves the layout to a tight version
plt.tight_layout()

plt.show()
```
