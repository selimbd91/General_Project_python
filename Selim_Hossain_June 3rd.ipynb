```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('Selim_Hossain_Data.csv')
df
```

```python
df.columns
```

```python
df.info()
```

```python
df.describe()
```

```python
#  i am taking temperature as my desired variable
print(df['Temperature(C)'])
mean_temp = df['Temperature(C)'].mean()
print(f'The average teperaure is = {mean_temp}')
```

```python
# median of the temperature 
temp_median = df['Temperature(C)'].median()
print(f'The temperature median is = {temp_median}')
```

```python
# mode of the temperature
temp_mode = df['Temperature(C)'].mode()
temp_mode
```

```python
# Applying the rolling mean using 2 data point as calculation window
temp_rolling_mean = df['Temperature(C)'].rolling(window= 2).mean()
print(f'The temperature rolling mean is = \n{temp_rolling_mean}')
```

```python
#Standard deviation of the temperature

temp_std = df['Temperature(C)'].std()
print(f'The temperature Standard Deviation is = {temp_std}')
```

```python
#The min() function returns the smallest of the input values.
temp_min = df['Temperature(C)'].min()
print(f'The minimum temperature is = {temp_min} degree')
```

```python
#similarly the max() function gives the largest value
temp_max = df['Temperature(C)'].max()
print(f'The maximum temperature is = {temp_max} degree')
```

```python
# describe() function gives the summery of statistical data
print(df.describe())
```

```python
# When having 'NaN's in your data array, nansum() is the better choice
print(f"The NaNsum of the temperature is = {np.nansum(df['Temperature(C)'])}")
```

```python
#numpy.cumsum() function is used when we want to compute the cumulative sum of array elements over a given axis.
df_cumsum = df[['Temperature(C)', 'DewPoint(F)', 'Humidity(Kg/m3)']]
df_cum = np.cumsum(df_cumsum, axis= 1)
print(f'The cumulative sum is: \n {df_cum}')
```

```python
#len() function gives the information about the length

df_temp = df[df['Temperature(C)'] >25]
print(f"The temperature large then 25 degree is= \n{df_temp['Temperature(C)']}")
print(f"The length of the temperature which is more than 25 degree is ={len(df_temp)}")
```

```python
# variance is a important statistical property

temp_var = df['Temperature(C)'].var()
print(f'The variance of the temperature is = {temp_var}')

```

```python
#covariance
#if your variable contains 'NaN', then this is the np.ma.cov function to solve the problem

covariance = np.ma.cov(df['Temperature(C)'], df['DewPoint(F)'])
print(f'The covariance between temperature and dewpoint is =\n{covariance}')
```

```python
sns.heatmap(covariance)
plt.show()
```

```python
#pivot_table

df_pivot_table = pd.pivot_table(df, values= 'Temperature(C)', index= ['Events'],aggfunc=np.sum, fill_value=0, columns='Transport')
df_pivot_table
```

```python
#Data visualization
#Histogram

ax = sns.distplot(df['Temperature(C)'], color='r')
ax.set_xlabel('Num of bins')
ax.set_ylabel('Frequency')
ax.set_title('Histogram graph')
plt.show()
```

```python
#Barplot is a categorical plot
ax = sns.barplot(x= df['Events'], y= df['Temperature(C)'], ci= None)
ax.set_title('Bar Plot')
```

```python
ax = sns.boxplot(data = [df['Temperature(C)'], df['DewPoint(F)'],df['Humidity(Kg/m3)']])
ax.legend(['Temperature', 'Dewpoint', 'Humidity'])
plt.show()
```

```python
plt.plot(df['EST'],df['Temperature(C)'],label= 'Temperature(C)')
plt.plot(df['EST'], df['DewPoint(F)'], label= 'Dewpoint(F)')
plt.legend()
plt.xlabel('Date')
plt.ylabel('Temperature and Dewpoint')
plt.title('Temperature and Dewpoint Graph')
plt.savefig('selim_assignment.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
#Scatter plot
ax = sns.scatterplot(data = df, x = 'Temperature(C)', y = 'DewPoint(F)', color= 'r')
ax.set_title('Scatter plot between Temperature and Dewpoint')
plt.show()
```
