```python
import pandas as pd
import seaborn as sns
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline

# https://ourworldindata.org/coronavirus-source-data
```

```python
ddff = pd.read_csv('owid-covid-data.csv')
ddff
```

```python
df = ddff[['date', 'location', 'new_cases', 'new_deaths','total_cases', 'total_deaths', 'cvd_death_rate']]
```

```python
df = df.iloc[0:29542, :]
dff = df.dropna()
```

```python
df['date'] = pd.to_datetime(df['date'])
df
```

## maximum cvd_death rate and country

```python
for i in df['location'].unique():
    max_of_cvd_death = df[df['location'] == i]['cvd_death_rate'].max()
    print('Maximum CVD_Death Rate in {a} is = {b:0.2f}'.format(a = i, b = max_of_cvd_death))     
```

```python
country_name = df[df['cvd_death_rate'] == df['cvd_death_rate'].max()].location.iloc[0]
cvd_death_rate = df['cvd_death_rate'].max()

print("The maximum CVD_Death Rate is = {a:0.2f} and the coutry is {b}".format(a = cvd_death_rate, b = country_name))
```

```python
plt.style.use('ggplot')
dff.astype({'new_cases': 'int32'})
plt.figure(figsize= (10, 5))
sns.distplot(dff['new_cases'])
plt.show()
```

## Average new case in different Countries 

```python
# mean function

for i in df['location'].unique():
    mean_of_newCase = df[df['location'] == i]['new_cases'].mean()
    print('Average new Cases in {a} is = {b:0.2f}'.format(a = i, b = mean_of_newCase))
```

```python
# median function

for i in df['location'].unique():
    median_of_newCase = df[df['location'] == i]['new_cases'].median()
    print('Median for new Cases in {a} is = {b:0.2f}'.format(a = i, b = median_of_newCase))
```

```python
# mode function

for i in df['location'].unique():
    mode_of_newCase = df[df['location'] == i]['new_cases'].mode()[0]
    print('Mode for new Cases in {a} is = {b:0.2f}'.format(a = i, b = mode_of_newCase))
```

## Total deaths in different country

```python
# sum function
for i in df['location'].unique():
    num_of_deaths = df[df['location'] == i]['total_deaths'].sum()
    print(f'Total deaths in {i} is = {num_of_deaths}')
```

```python
# max function
max_death = df['total_deaths'].max()
country_max = df[df['total_deaths'] == df['total_deaths'].max()]['location'].iloc[0]
max_death

print(f"The maximum deaths is = {max_death} and \nThe Country is {country_max}.")
```

```python
plt.figure(figsize= (10, 5))
sns.lineplot(x = df['date'], y = df['new_cases'], ci = None)
plt.title('Worldwide New Cases with Time', fontsize= 25)
plt.xlabel('Time', fontsize= 20)
plt.ylabel('New_Cases', fontsize= 20)
plt.show()
```

```python
plt.figure(figsize= (10, 5))
sns.lineplot(x = df['date'], y = df['new_deaths'], ci = None)
plt.title('Worldwide new_deaths with Time', fontsize= 25)
plt.xlabel('Time', fontsize= 20)
plt.ylabel('New_Deaths', fontsize= 20)
plt.show()
```

## Determining Rolling Statistics for New Deaths in Germany

```python
german_df = df[df['location'] == 'Germany']
china_df = df[df['location'] == 'China']
bangladesh_df = df[df['location'] == 'Bangladesh']
india_df = df[df['location'] == 'India']
spain_df = df[df['location'] == 'Spain']
```

```python
rol_mean = german_df['new_deaths'].rolling(window= 12).mean()
rol_std = german_df['new_deaths'].rolling(window= 12).std()
plt.figure(figsize= (10, 5))
sns.lineplot(x = df['date'], y = german_df['new_deaths'], ci = None, label= 'Original')
sns.lineplot(x = df['date'], y= rol_mean, ci= None, label = 'Rolling Mean')
sns.lineplot(x = df['date'], y= rol_std, ci= None, label = 'Rolling Std')
plt.title('Rolling mean and Standard Devition in Germany', fontsize= 25)
plt.xlabel('Time', fontsize= 20)
plt.ylabel('New_Deaths', fontsize= 20)
plt.legend()
plt.show()
```

```python
plt.figure(figsize= (10, 5))
box_det = [german_df['total_deaths'], china_df['total_deaths'], bangladesh_df['total_deaths'], india_df['total_deaths'], spain_df['total_deaths']]
sns.boxplot(data= box_det, width=0.5, notch = True)
plt.xticks([0, 1, 2, 3, 4], ['Germany', 'China', 'BD', 'Indida', 'Spain'])
plt.title('Boxplot for Five Countries', fontsize= 25)
plt.show()
```

```python
#Pie plot.Normally use for categorical data
plt.figure(figsize= (10, 5))
labels = ['Germany', 'China', 'Bangladesh', 'India', 'Spain']
sizes = [german_df['new_cases'].sum(), china_df['new_cases'].sum(), bangladesh_df['new_cases'].sum(), india_df['new_cases'].sum(), spain_df['new_cases'].sum()]

plt.pie(sizes, labels=labels,autopct='%1.1f%%',shadow=True, startangle=90)
plt.title('Five Countries New CVD cases ', fontsize= 25)
plt.show()
```

# Confirmed cases throughout the world

```python
daily_cases=df.groupby("date").total_cases.sum()
daily_cases=daily_cases.to_frame().reset_index()
daily_cases.rename(columns={'total_cases':'cum_cases'}, inplace=True)
daily_cases.head(3)
```

```python
plt.style.use('ggplot')
plt.figure(figsize= (10, 5))
plt.plot(daily_cases.date,daily_cases.cum_cases,'r')

plt.xticks(rotation=45)
plt.ylabel("Accumulate Cases Worldwide")
plt.title('Worldwide confirmed cases')

plt.show();
```

# China and Germany compared to the rest of the world

```python
df['ISChinaGermany']=df['location'].apply(lambda x: "China" if x=="China" else ('Germany' if x=='Germany' else 'others'))
df.head(3)
```

```python
ChinaGermany_vs_others=df.groupby(["date","ISChinaGermany"]).total_cases.sum().reset_index()
ChinaGermany_vs_others
```

```python
plt.figure(figsize= (10, 5))
ChinaGermany_vs_others.set_index('date').groupby('ISChinaGermany')['total_cases'].plot(style='--', legend=True,rot=45, ylim=(0,2000000));
plt.title('China,Germany and Other coutries total cases with time', fontsize= 20)
```

```python
df['active'] = df['total_cases']
top = df[df['date']==df['date'].max()]
top_actives = top.groupby(by = 'location')['active'].sum().sort_values(ascending= False).head(20).reset_index()
plt.figure(figsize= (15, 10))
plt.xticks(fontsize= 15)
plt.yticks(fontsize = 15)
plt.xlabel('Total cases', fontsize = 25)
plt.ylabel('Country', fontsize= 25)
plt.title('Top 20 countries having most active cases', fontsize= 25)
ax = sns.barplot(x = top_actives.active, y= top_actives.location)
for i, (value, name) in enumerate(zip(top_actives.active, top_actives.location)):
    ax.text(value, i-.05, f'{value:,.0f}', size= 10, ha= 'left', va= 'center')
ax.set(xlabel= 'Total cases', ylabel = 'Country')
plt.show()
```
